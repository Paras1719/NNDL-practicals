```python
import numpy as np
```

```python
A=np.array([[2,3],[5,5]])
B=np.array([[4,6],[2,3]])

print("Matrix Addition:\n", A+B)
print("Matrix Multiplication:\n", np.matmul(A,B))
```

```python
C = np.array([[2, 3, 1],
              [5, 5, 2],
              [4, 1, 3]])
D = np.array([[4, 6, 2],
              [2, 3, 7],
              [1, 0, 5]])

print("Matrix Multiplication:\n", np.matmul(C,D))
```
